```python
students_marks = {
    'Alice': 85,
    'Bob': 92,
    'Charlie': 78,
    'David': 65,
    'Eve': 90
}

print("Student Marks Dictionary:")
print(students_marks)
```

```python
total_marks = sum(students_marks.values())
num_students = len(students_marks)

average_marks = total_marks / num_students

print(f"Total marks: {total_marks}")
print(f"Number of students: {num_students}")
print(f"Average marks: {average_marks:.2f}")
```

```python
# Find the student with the highest mark
highest_scorer = max(students_marks, key=students_marks.get)
highest_mark = students_marks[highest_scorer]

# Find the student with the lowest mark
lowest_scorer = min(students_marks, key=students_marks.get)
lowest_mark = students_marks[lowest_scorer]

print(f"\nHighest Scorer: {highest_scorer} with {highest_mark} marks")
print(f"Lowest Scorer: {lowest_scorer} with {lowest_mark} marks")
```

```python
print(f"\nStudents scoring above the average ({average_marks:.2f}):")
for student, marks in students_marks.items():
    if marks > average_marks:
        print(f"- {student}: {marks} marks")
```

```python
student_name_search = input("Enter the name of the student you want to search for: ")

if student_name_search in students_marks:
    print(f"\n{student_name_search} found! Marks: {students_marks[student_name_search]}")
else:
    print(f"\n{student_name_search} not found in the student records.")
```

```python
student_to_update = input("Enter the name of the student whose marks you want to update: ")

if student_to_update in students_marks:
    try:
        new_mark = int(input(f"Enter the new mark for {student_to_update}: "))
        if 0 <= new_mark <= 100:
            students_marks[student_to_update] = new_mark
            print(f"\n{student_to_update}'s marks updated to {new_mark}.")

            # Recalculate and display the revised average
            total_marks = sum(students_marks.values())
            num_students = len(students_marks)
            average_marks = total_marks / num_students
            print(f"Revised Student Marks Dictionary: {students_marks}")
            print(f"Revised Average Marks: {average_marks:.2f}")
        else:
            print("Mark must be between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a number for the mark.")
else:
    print(f"\n{student_to_update} not found in the student records. Marks not updated.")
```
